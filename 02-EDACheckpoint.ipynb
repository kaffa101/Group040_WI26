# Research Question

Is there a relationship between the COGS course drop rate at UCSD and students' evaluations of these courses - specifically course workload (study hours per week), instructor ratings, and average course grades - using CAPE evaluation data from 2022-23 school year (FA22 - SP23), on COGS1, 14A, 17, 101B, and 107A?

## Background and Prior Work

It's so common for college students to choose to drop their courses within the beginning few weeks of a new quarter, that we often overlook the significance behind this behavior: what factors could actually be correlated with this complex decision - giving up a course that is planned or even anticipated for months since registration? Specifically, aside from non-academic reasons, what aspects of a course itself may be associated with a high withdrawal rate?

One previous research has been conducted on the affective reasoning of why students might choose to withdraw courses.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Among the multiple reasons for dropping from a course, "at risk of failing the course" is ranked at the top with a 39.9% probability of withdrawing. In other words, students may tend to drop courses if they feel they may get bad grades, mainly due to the difficulty, as the courses continue. 

Another research has pointed out that "assignment too heavy" could be a reason of dropping courses as well.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) This term, first of all, may be regarded as an implication of the difficulty of the course. However, this research was conducted through community colleges, and academic workload is related to job conflict (time conflict) specifically in this case. Therefore, heavy assignments also represent time consuming tasks. 

We also discovered a prior study that takes the teaching quality into account.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) By examining SRT (students rating of teaching) data, the researchers show that students can have their own evaluations of the instructors only after taking a few classes, which is then a critical factor of if the students will drop the course. Noteworthy, this factor is isolated from the probability of receiving bad grades or being assigned heavy tasks, because teaching methods or attitudes of professors do not necessarily relate to their assignments or rating scales.

Drawing on the findings of the researches above, we include three major instructional factors - average course grades, course workload (work time), and instructor ratings - in our research of their correlations to the students' dropping rate of COGS courses at UCSD, examining if they show the same correlations in our case as well.

"Enrolled" displays integer numbers of the currently enrolled number of people to the course over time. "Waitlisted" displays the integer numbers of the currently waitlisted number of people to the course over time. We will monitor their changes from the beginning of the quarter to the end of week 4 - when the "drop" action ends. Everytime the "waitlisted" number decreases by 1 from the beginning of instruction to the end of week 2 (the last day of waitlist), we count it as 1 person dropping, because that's when waitlist automatically fills the enrollment in once a spot is available. From week 3 to the end of week 4, everytime the "enrolled" number decreases by 1, we count it as 1 person dropping, because waitlist has already ended and people only drop out from the course.
  - Shortcomings: Since the dataset only provides the changes of enrollment and waitlist number over the time, but not directly showing when the "drop" action happens, we can only infer drop data from the changes of enrollment and waitlist, which may not be too precise. We need to assume that most of people stay on the waitlist for the first two weeks.

```python
import pandas as pd
import os 
def drop_rate(course):
    diff = course['enrolled'].diff().fillna(0)
    total_enroll = int(diff[diff > 0].sum())
    total_drop = int((-diff[diff < 0]).sum())
    if total_enroll == 0:
        drop_rate = 0
    else:
        drop_rate = round((total_drop / total_enroll) * 100, 2)
    return total_enroll, total_drop, drop_rate

def quarter_to_df(file):
    results = []
    for filename in os.listdir(file):
        if filename.endswith(".csv"):
            file_path = os.path.join(file, filename)
            df = pd.read_csv(file_path)
            enroll, drop, rate = drop_rate(df)
            course_name = filename.replace(".csv", "")
            results.append({"course": course_name, "total_enrolled": enroll, \
                "total_dropped": drop, "drop_rate_percent": rate})
    results_df = pd.DataFrame(results)
    results_df = results_df.set_index("course")
    return results_df
```

### Dataset #2 
#### Datasets Merging
We have the FA22 - SP23 COGS courses enrollment data in our raw data folder. However, they are messy, individual csv files, and lack proper identifications indicating their quarter. Here, we need to merge all these datasets into one big csv file, also adding two new columns "class" and "quarter" to identify the data in the new file. The merged file will be saved in the interim folder.

```python
import os
import glob

quarter_folders = {
    'FA22': 'data/00-raw/COGS_22FA',
    'WI23': 'data/00-raw/COGS_23WI',
    'SP23': 'data/00-raw/COGS_23SP',
}
df_list = []
for quarter, folder in quarter_folders.items():
    pattern = os.path.join(folder, '*.csv')
    for path in glob.glob(pattern):
        df = pd.read_csv(path)
        filename = os.path.basename(path)          
        class_name = os.path.splitext(filename)[0]   
        df['class'] = class_name
        df['quarter'] = quarter
        df_list.append(df)

merged_df = pd.concat(df_list, ignore_index=True)
merged_df.to_csv('data/01-interim/merged_cogs.csv', index = False)
merged_df.head()
```

#### Data Cleaning
Now, we have the merged csv file, but it is surprisingly huge. Although there's no missing value due to the nature of the enrollment data, because the file records every 10 mins since the first day to course enrollment, we actually don't need most of the data and should only keep the parts from the beginning of a new quarter's instruction to the end of week 4 of that quarter (the course withdrawl happening before a quarter starts should not be counted as an action related to the course itself). Therefore, we need to delete the undesired data from the file and only keep data from these three time ranges: 2022.09.22 - 2022.10.21, 2023.01.09 - 2023.02.03, and 2023.04.03 - 2023.04.28. We will save the new csv file in the processed data folder.

```python
merged_df['time'] = pd.to_datetime(merged_df['time'])
time_ranges = [
    ('2022-09-22', '2022-10-21'),
    ('2023-01-09', '2023-02-03'),
    ('2023-04-03', '2023-04-28'),
]
ranges = []
for i, (start, end) in enumerate(time_ranges, start = 1):
    df_ranges = merged_df[(merged_df['time'] >= start) & (merged_df['time'] <= end)].copy()
    ranges.append(df_ranges)
clean_enroll = pd.concat(ranges, ignore_index = True)
clean_enroll.to_csv('data/02-processed/enroll_clean.csv')
clean_enroll.shape
```

### Exploratory Data Analysis

In this section, we analyze whether COGS course drop rate is associated with three course-evaluation factors from CAPE data: workload (`Study Hours per Week`), instructor rating (`Percentage Recommended Professor`), and performance (`Average Grade Received`).

To keep this section reproducible and fast to run, we directly load the processed files from `data/02-processed`:
- `capes_clean.csv` (evaluation features)
- `enroll_clean.csv` (time-series enrollment snapshots)

We first build a course-quarter drop-rate table from enrollment changes, merge it with CAPE variables, and then examine distributions and pairwise relationships using visual EDA.

**Drop-rate definition used in EDA:**
\[
\text{drop\_rate} = \frac{\text{total\_drop}}{\text{starting\_enrolled} + \text{total\_add}} \times 100
\]
This avoids impossible values above 100% caused by using only within-window adds in the denominator.

#### Section 1: Build Analysis Table + Univariate Distributions

This section constructs the final EDA table at the `class x quarter` level and checks distributions/outliers for the variables used in our research question.

##### Initialization

```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', context='notebook', palette='deep')

# load the processed datasets
capes = pd.read_csv('data/02-processed/capes_clean.csv')
enroll = pd.read_csv('data/02-processed/enroll_clean.csv')
```

##### Helper functions

```python
def parse_percentage(value):
    if pd.isna(value):
        return np.nan
    return float(str(value).replace('%', '').strip())

def parse_gpa(value):
    if pd.isna(value):
        return np.nan
    match = re.search(r'\((\d+(?:\.\d+)?)\)', str(value))
    return float(match.group(1)) if match else np.nan

def extract_class_name(course_text):
    if pd.isna(course_text):
        return np.nan
    match = re.search(r'(COGS\s+\d+[A-Z]?)', str(course_text))
    return match.group(1) if match else np.nan

def compute_drop_metrics(group):
    # Enrollment snapshots can have many add/drop events. Using
    # total_drop / total_add can exceed 100% when a class starts with
    # non-zero enrollment before our observation window.
    #
    # We approximate drop rate as:
    #   total_drop / (starting_enrolled + total_add)
    # which is bounded in [0, 100] for this dataset and better reflects
    # the proportion of enrollment opportunities that ended in drops.
    diff = group['enrolled'].diff().fillna(0)
    total_add = int(diff[diff > 0].sum())
    total_drop = int((-diff[diff < 0]).sum())
    starting_enrolled = int(group['enrolled'].iloc[0]) if len(group) else 0
    total_exposure = starting_enrolled + total_add

    if total_exposure == 0:
        drop_rate = 0.0
    else:
        drop_rate = round((total_drop / total_exposure) * 100, 2)

    return pd.Series({
        'starting_enrolled': starting_enrolled,
        'total_add': total_add,
        'total_drop': total_drop,
        'total_exposure': total_exposure,
        'drop_rate': drop_rate,
        'n_snapshots': len(group)
    })
```

##### Process data for plotting


```python
enroll['time'] = pd.to_datetime(enroll['time'])
enroll = enroll.sort_values(['class', 'quarter', 'time']).copy()

drop_df = (
    enroll.groupby(['class', 'quarter'])
    .apply(compute_drop_metrics, include_groups=False)
    .reset_index()
)

capes = capes.copy()
capes['class'] = capes['Course'].apply(extract_class_name)
capes['quarter'] = capes['Quarter'].astype(str).str.strip().str.upper()
capes['prof_recommend_pct'] = capes['Percentage Recommended Professor'].apply(parse_percentage)
capes['grade_gpa'] = capes['Average Grade Received'].apply(parse_gpa)
capes['study_hours'] = pd.to_numeric(capes['Study Hours per Week'], errors='coerce')

eda_df = capes.merge(drop_df, on=['class', 'quarter'], how='inner')
analysis_cols = ['drop_rate', 'study_hours', 'prof_recommend_pct', 'grade_gpa']
eda_df = eda_df.dropna(subset=analysis_cols).copy()

print('Rows in final EDA table:', len(eda_df))
print('Unique course-quarter combinations:', eda_df[['class', 'quarter']].drop_duplicates().shape[0])
print('Max drop rate:', eda_df['drop_rate'].max())
display(eda_df[['class', 'quarter'] + analysis_cols].head(10))
display(eda_df[analysis_cols].describe().T)
display(eda_df.head())
```

##### Univariate

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
plot_map = [
    ('drop_rate', 'Drop Rate (%)'),
    ('study_hours', 'Study Hours / Week'),
    ('prof_recommend_pct', 'Professor Recommendation (%)'),
    ('grade_gpa', 'Average Grade GPA')
]
hist_colors = sns.color_palette('Set2', 4)

for ax, (col, title), color in zip(axes.flatten(), plot_map, hist_colors):
    sns.histplot(data=eda_df, x=col, bins=15, kde=True, ax=ax, color=color, edgecolor='white', alpha=0.9)
    ax.set_title(title)
    ax.set_xlabel(title)
    ax.set_ylabel('Count')

plt.tight_layout()
plt.show()

# Zoom the quarter boxplot so outliers do not squash the boxes
plt.figure(figsize=(8, 4))
quarter_order = sorted(eda_df['quarter'].dropna().unique())
q1, q3 = eda_df['drop_rate'].quantile([0.25, 0.75])
iqr = q3 - q1
upper_fence = q3 + 1.5 * iqr
visible_df = eda_df[eda_df['drop_rate'] <= upper_fence].copy()
outlier_count = int((eda_df['drop_rate'] > upper_fence).sum())

sns.boxplot(
    data=visible_df,
    x='quarter',
    y='drop_rate',
    hue='quarter',
    order=quarter_order,
    palette='Set1',
    legend=False,
    showfliers=False
)
sns.stripplot(
    data=visible_df,
    x='quarter',
    y='drop_rate',
    order=quarter_order,
    color='black',
    alpha=0.35,
    size=3
)

plt.title(f'Drop Rate Distribution by Quarter (outliers hidden: {outlier_count})')
plt.xlabel('Quarter')
plt.ylabel('Drop Rate (%)')
plt.tight_layout()
plt.show()

# Full-range version (includes all points/outliers)
plt.figure(figsize=(8, 4))
sns.boxplot(
    data=eda_df,
    x='quarter',
    y='drop_rate',
    hue='quarter',
    order=quarter_order,
    palette='Set1',
    legend=False,
    showfliers=True
)
sns.stripplot(
    data=eda_df,
    x='quarter',
    y='drop_rate',
    order=quarter_order,
    color='black',
    alpha=0.25,
    size=3
)
plt.title('Drop Rate Distribution by Quarter (full range)')
plt.xlabel('Quarter')
plt.ylabel('Drop Rate (%)')
plt.tight_layout()
plt.show()
```

#### Section 2: Relationships Between Drop Rate and Course Factors

Here we visualize how drop rate changes with workload, recommendation rate, and average grade, then compare low-drop vs high-drop courses using quartile-based groups.

##### Bivariate Relationships

```python
features = [
    ('study_hours', 'Study Hours / Week'),
    ('prof_recommend_pct', 'Professor Recommendation (%)'),
    ('grade_gpa', 'Average Grade GPA')
]
reg_colors = sns.color_palette('Dark2', 3)

fig, axes = plt.subplots(1, 3, figsize=(16, 4), sharey=True)
for ax, (xcol, xlabel), color in zip(axes, features, reg_colors):
    sns.regplot(
        data=eda_df,
        x=xcol,
        y='drop_rate',
        color=color,
        scatter_kws={'alpha': 0.75, 's': 45},
        line_kws={'linestyle': '--', 'linewidth': 2},
        ci=None,
        ax=ax
    )
    ax.set_xlabel(xlabel)
    ax.set_title(f'Drop Rate vs {xlabel}')

axes[0].set_ylabel('Drop Rate (%)')
plt.tight_layout()
plt.show()
```

##### Quartile-based comparison

```python
q1 = eda_df['drop_rate'].quantile(0.25)
q3 = eda_df['drop_rate'].quantile(0.75)

low_drop = eda_df[eda_df['drop_rate'] <= q1].copy()
high_drop = eda_df[eda_df['drop_rate'] >= q3].copy()
low_drop['group'] = 'Low drop (Q1)'
high_drop['group'] = 'High drop (Q4)'
compare_df = pd.concat([low_drop, high_drop], ignore_index=True)

comparison_cols = ['study_hours', 'prof_recommend_pct', 'grade_gpa']
comparison_labels = ['Study Hours', 'Prof Recommendation (%)', 'Grade GPA']
boxplot_palettes = [
    {'Low drop (Q1)': '#66c2a5', 'High drop (Q4)': '#498b76ff'},
    {'Low drop (Q1)': '#8da0cb', 'High drop (Q4)': '#6e7c9e'},
    {'Low drop (Q1)': '#a6d854', 'High drop (Q4)': '#536c2b'}
]

fig, axes = plt.subplots(
    1, 3,
    figsize=(18, 5.4),
    constrained_layout=True
)
for ax, col, label, pal in zip(axes, comparison_cols, comparison_labels, boxplot_palettes):
    sns.boxplot(data=compare_df, x='group', y=col, hue='group', palette=pal, legend=False, ax=ax)
    sns.stripplot(data=compare_df, x='group', y=col, color='black', alpha=0.35, size=3, ax=ax)
    ax.set_title(label)
    ax.set_xlabel('')

    vals = compare_df[col].dropna()
    if col == 'prof_recommend_pct':
        low_q, high_q = vals.quantile([0.10, 0.995])
    else:
        low_q, high_q = vals.quantile([0.02, 0.98])
    span = high_q - low_q
    pad = span * 0.12 if span > 0 else 0.1
    ax.set_ylim(low_q - pad, high_q + pad)

fig.suptitle('Low vs High Drop-Rate Courses')
plt.show()

summary = pd.DataFrame({
    'low_drop_median': low_drop[comparison_cols].median(),
    'high_drop_median': high_drop[comparison_cols].median(),
})
summary['difference_high_minus_low'] = summary['high_drop_median'] - summary['low_drop_median']
display(summary.round(3))
```
